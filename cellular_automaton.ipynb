```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
def game_play(roles:dict, initial_cell, iteritons=100):
    cells_num = len(initial_cell)
    game_arr = np.zeros((iteritons+1, cells_num))
    game_arr[0] = initial_cell
    
    for i in range(1, iteritons+1):
        holder = game_arr[i-1]
        for j in range(cells_num):
            if j == cells_num-1:
                s[j] = roles[(holder[j-1], holder[j], holder[0])]
            else:
                s[j] = roles[(holder[j-1], holder[j], holder[j+1])]
        game_arr[i] = s
    return game_arr
```

```python
def rule_num(num):
    num = np.binary_repr(num, width=8)
    return {
    (1, 1, 1): int(num[0]),
    (1, 1, 0): int(num[1]),
    (1, 0, 1): int(num[2]),
    (1, 0, 0): int(num[3]),
    (0, 1, 1): int(num[4]),
    (0, 1, 0): int(num[5]),
    (0, 0, 1): int(num[6]),
    (0, 0, 0): int(num[7]), 
}
```

```python
s = np.zeros(100)
s[49] = 1
```

```python
rule_30 = game_play(rule_num(41), s)
```

```python
plt.imshow(rule_30, cmap='Greys')
```

```python
def random_road(p, cell_len=100):
    cars_num = round(p*cell_len)
    initial_road = np.zeros(cell_len)
    initial_road[:cars_num] = 1
    np.random.shuffle(initial_road)
    return game_play(rule_num(184), initial_road)
```

```python
k = random_road(0.01)
plt.imshow(k)
```

```python
def avg_speed(road):
    v_arr = []
    for i in range(1, len(road)):
        sub = np.subtract(road[i], road[i-1])
        m_t = np.count_nonzero(sub == -1)
        v = m_t/np.count_nonzero(road[i])
        v_arr.append(v)
    return sum(v_arr)/(len(road))
```

```python
avg_speed(random_road(0.4))
```

```python
p_arr = []
v_avg = []
j = []
for k in np.arange(0.01, 1, 0.01):
    traffic = random_road(k)
    v = avg_speed(traffic)
    p_arr.append(k)
    v_avg.append(v)
    j.append(v*k)
```

```python
plt.plot(p_arr, v_avg, 'b')
```

```python
plt.plot(p_arr, j, 'b')
```
